#### **Задание 1. Работа с тензорами**
1. Создайте два тензора: один с равномерным распределением значений, другой — с нормальным.
2. Выполните между ними элементные арифметические операции: сложение, умножение.
3. Преобразуйте результат в массив NumPy и выведите его.

```python
# 1. Создаем два тензора
tensor_uniform = torch.rand((3, 3))  # равномерное распределение на [0, 1)
tensor_normal = torch.randn((3, 3))  # нормальное распределение N(0, 1)

print("Тензор с равномерным распределением:\n", tensor_uniform)
print("Тензор с нормальным распределением:\n", tensor_normal)

# 2. Элементные операции
sum_result = tensor_uniform + tensor_normal
mul_result = tensor_uniform * tensor_normal

print("\nСумма тензоров:\n", sum_result)
print("\nЭлементное умножение:\n", mul_result)

# 3. Преобразование в NumPy
sum_numpy = sum_result.numpy()
print("\nРезультат суммы в формате NumPy:\n", sum_numpy)
```
